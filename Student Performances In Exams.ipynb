```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#setting the pass mark
passmark = 50
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#checking null values
df.isnull().sum()
```

```python
#checking the correlation between columns
sns.heatmap(df.corr(), cmap = 'coolwarm')
```

```python
#distribution plot for the maths score
sns.distplot(df['math score'],
            bins = 50)
```

```python
#distribution plot for the reading score
sns.distplot(df['reading score'],
            bins = 50)
```

```python
#distribution plot for the writing score
sns.distplot(df['writing score'],
            bins = 50)
```

```python
#Checking the amount of people who passed the exams using countplot
df['Math_Pass'] = np.where(df['math score'] < passmark, 'F', 'P')
df.Math_Pass.value_counts()
```

```python
#Researching whether intelligence can be passed around generations
sns.countplot(x = 'parental level of education', 
              data = df, 
              hue = 'Math_Pass', 
              palette = 'coolwarm')
plt.xticks(rotation = 90);
```

```python
df['Reading_Pass'] = np.where(df['reading score'] < passmark, 'F', 'P')
df.Reading_Pass.value_counts()
```

```python
sns.countplot(x = 'parental level of education', 
              data = df, 
              hue = 'Reading_Pass', 
              palette = 'coolwarm')
plt.xticks(rotation = 90);
```

```python
df['Writing_Pass'] = np.where(df['writing score'] < passmark, 'F', 'P')
df.Writing_Pass.value_counts()
```

```python
sns.countplot(x = 'parental level of education', 
              data = df, 
              hue = 'Writing_Pass', 
              palette = 'coolwarm')
plt.xticks(rotation = 90);
```
